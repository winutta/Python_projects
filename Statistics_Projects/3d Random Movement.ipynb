```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, sem, uniform, chi2
from mpl_toolkits import mplot3d
```

```python
rv1 = norm.rvs(size = 1000)
rv12 = norm.rvs(size = 1000)
rv2 = uniform.rvs(size = 1000)
rv3 = chi2.rvs(size = 1000,df = 2)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(rv1,rv2,rv3)
plt.show()
```

```python
fig = plt.subplots()
for i in range(len(rv1)):
    ax.scatter(rv1[i],rv2[i])
    #plt.pause(0.5)
    plt.draw()
plt.show()
```
