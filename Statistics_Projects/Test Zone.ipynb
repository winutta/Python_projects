```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

fig,ax = plt.subplots(1,1)

# Plot the absolute probability density function along 100 
# evenly spaced point from the 1% and 99% percentage point values
# on the norm dist, just using the pdf() function
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99),100)
ax.plot(x,norm.pdf(x),'r-',lw=5,alpha=0.6, label= 'normal dist prob dens f')

#Plot the 'frozen' pdf along those same points by just using a frozen norm()

rv = norm()
ax.plot(x,rv.pdf(x),'k-',lw=2,label='frozen pdf')

#plot a histogram of 1000 points on a normal distribution to 'fill in'
#the graph. This is done using the rvs() function
r = norm.rvs(size=1000)
ax.hist(r,density = True, histtype = 'stepfilled', alpha=0.2)
ax.legend(loc='best',frameon=False)
plt.show()
```
