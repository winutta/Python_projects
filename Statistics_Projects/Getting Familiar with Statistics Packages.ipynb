```python
from scipy import stats
```

```python
fig,ax = plt.subplots(1,1)

x = np.linspace(stats.norm.ppf(0.01),stats.norm.ppf(0.99),100)
ax.plot(x,stats.norm.pdf(x),'r-',lw=5,alpha=0.6,label='norm pdf')

rv = stats.norm()
ax.plot(x,rv.pdf(x),'k-',lw=2,label='frozen pdf')
vals = stats.norm.ppf([0.001, 0.5, 0.999])
print(np.allclose([0.001, 0.5, 0.999], stats.norm.cdf(vals)))

r = stats.norm.rvs(size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2,label='histogram')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from scipy.stats import chi2

fig2,ax2 = plt.subplots(1,1)
df = 6
x2 = np.linspace(chi2.ppf(0.01,df),chi2.ppf(0.99,df),100)

ax2.plot(x2,chi2.pdf(x2,df),'r-',lw=5,label = 'chi2 pdf')

rv2 = chi2(df)
ax2.plot(x2,rv2.pdf(x2),'k-',lw=2,label='frozen chi2')

r2 = chi2.rvs(df,size=1000)
ax2.hist(r2,density=True, histtype = 'stepfilled',alpha=0.2,label = 'hist')
ax2.legend(loc = 'best',frameon=False)
plt.show()
```

```python
def distplot(pkg):
    import matplotlib.pyplot as plt
    import importlib
    lib = "scipy.stats."+pkg
    plugin_lib = importlib.import_module(lib,".")
    import plugin_lib as distr
    fig,ax = plt.subplots(1,1)
    
    x = linspace(distr.ppf(0.01),distr.ppf(0.99),100)
    ax.plot(x,distr())
```
