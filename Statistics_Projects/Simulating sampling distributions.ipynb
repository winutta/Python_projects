```python
from scipy.stats import norm, sem
import matplotlib.pyplot as plt
import numpy as np
```

```python
rv = norm.rvs(size = 10000)

print("Mean of 10,000 randomly sampled data points: ", np.mean(rv))
print("Std Dev. of the data points: ",np.std(rv))
print("Function result for standard error of the mean", sem(rv))
```

```python
print("95% confidence interval [",np.mean(rv)-1.96*sem(rv),",",np.mean(rv)+1.96*sem(rv),"]")
```

Therefore we cannot reject that the mean is anywhere in the confidence interval at this confidence level.

```python
#comparison with function found in online forum https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data

# we are spot on!

import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

mean_confidence_interval(rv,confidence = 0.95)
```

```python
# Lets try with 2 groups

a = rv[:int(len(rv)/2)]
b = rv[int(len(rv)/2):]
n = len(rv)
ma, sea = np.mean(a),scipy.stats.sem(a)
mb, seb = np.mean(b),scipy.stats.sem(b)
ha = sea * 1.96
hb = seb * 1.96
print("mean of a: ", ma, "\n95% Confidence Interval [",ma-ha,",",ma+ha,"]\n")
print("mean of b: ", mb, "\n95% Confidence Interval [",mb-hb,",",mb+hb,"]")
```

```python
def make_partition(r,c):
    if len(r)/c%1!=0.0:
        print("sorry not divisible")
        return
    else:
       
        part = {
        }
        for i in range(int(len(r)/c)):
            part[i] = r[:]
        
```

```python
def chunkIt(seq, num):
    if len(seq)/num%1!=0.0:
        print("sorry not divisible")
        return
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

```python
def get_stuff(v):
    if v is None:
        return
    li = []
    for i in range(len(v)):
        li.append(np.mean(v[i]))
    m = np.mean(li)
    s = np.std(li)
    return m,s

```

```python
def test_a_bunch(s):
    ind = []
    mns = []
    std = []
    for i in range(len(s)):
        temp = get_stuff(chunkIt(s,i+1))
        print("Trying to comlete ",i+1,"\: ")
        print("mean, std = ",temp)
        if temp is not None:
            ind.append(i)
            mns.append(temp[0])
            std.append(temp[1])
    return ind, mns, std
```

# Very import point. this is only for samples/partitions of equal size. This does not represent the total distribution of samples. I need to instead take random samples of size n without replacement over and over until I get a distribution.

```python
ind, mns, std = test_a_bunch(rv)
```
